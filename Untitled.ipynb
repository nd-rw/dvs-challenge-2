```python
import pandas as pd
import os
import re
import math
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from functools import reduce
import matplotlib.pyplot as plt
```

```python
sid = SentimentIntensityAnalyzer()
ss = sid.polarity_scores
```

```python
def create_emoji_dict():
    """
    Uses a combination of code from two functions from the VaderSentiment.py file
    from the python library Vader to convert emoji lexicon file to a dictionary.
    """
    emoji_full_filepath = 'dvs2_data/emoji_utf8_lexicon.txt'
    with open(emoji_full_filepath, encoding='utf-8') as f:
        emoji_full_filepath = f.read()
    emoji_dict = {}
    for line in emoji_full_filepath.split('\n'):
        (emoji, description) = line.strip().split('\t')[0:2]
        emoji_dict[emoji] = description
    return emoji_dict
```

```python
emojis_df = pd.read_csv('dvs2_data/emoji_over_time.csv')
emojis = set(emojis_df['emoji'])
emoji_dict = create_emoji_dict()
```

```python
def cleanEmojiString(emoji):
    stringsToSkip = ['-1', '123', '1234', 'd3']
    charReplacements = {'-' : ' ',
                        '_' : ' ',
                        ':' : '',
                       'skin tone' : '',
                       '2' : '',
                       '3' : '',
                       '4' : '',
                       '5' : ''}
    if any(text in emoji for text in stringsToSkip):
        print(emoji)
        return emoji
    elif any(char in emoji for char in charReplacements.keys()):
        cleaned_emoji_string = reduce(lambda a, kv: a.replace(*kv), charReplacements.items(), emoji)
        if cleaned_emoji_string[-1] == " ":
            return cleaned_emoji_string[:-1]
        else:
            return cleaned_emoji_string
    else:
        return emoji
```

```python
def emojiSentiment(emoji_clean):
    sentiment_value = ss(emoji_clean)['compound']
    return sentiment_value
```

```python
def thumbReplacement(thumb_string):
    thumbs = {'-1' : 'rejection', '+1' : 'satisfaction'}
    # used the words replacement words 'rejection' and 'approval'
    # as Vader returns a similar compound sentiment score for both.
    
    if any(thumb in thumb_string for thumb in thumbs.keys()):
        return thumbs[thumb_string]
    else:
        return thumb_string
```

```python
emojis_df['emoji_clean'] = emojis_df.emoji.apply(cleanEmojiString)
```

```python
emojis_df['emoji_clean'] = emojis_df.emoji_clean.apply(thumbReplacement)
```

```python
emojis_df['emoji_score'] = emojis_df.emoji_clean.apply(emojiSentiment)
emojis_df['overall_sentiment'] = emojis_df['emoji_score'] * emojis_df['value']
```
